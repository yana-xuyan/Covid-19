```python
import json
import pickle
import pprint 

pp = pprint.PrettyPrinter(width=200)
```

```python
DATA_PATH = "../test/kaggel_task_2.pt"

with open(DATA_PATH, "rb") as reader:
    data = pickle.load(reader)

# List [
#     "question": "xxxx",
#     "data": [
#         {
#             "answer": "xxxx",
#             "context": "xxxx",
#         }
#     ]
# ] 
# =>
# {
#     "version": "",
#     "data": [
#         {
#             "title":"",
#             "paragraphs": [
#                 {
#                     "context": ""
#                     "qas": [
#                         {
#                             'question': "",
#                             'id': "",
#                             'answer': "",
#                         }
#                     ]
#                 }
#             ]
#         }
#     ]
# }
```

```python
synthetic = {
    "version": "synthetic-task2",
    "data": []
}

synthetic = {
    'version': 'synthetic',
    'data': []
            }

data_item = {
    'title': 'synthetic',
    'paragraphs': []
}

paragraph_item = {
    'context': '',
    'qas': []
}

# qas_item = {'question': '', 'id': ''}
```

```python
idx = 0
for data_item in data:
    question = data_item["question"]
    qas = data_item["data"]
    for qa in qas:
        context = qa["context"]
        answer = qa["answer"]
        
        paragraph_item = {
            'context': context,
            'qas': []
        }

        qas_item = {
            'question': question, 
            'id': idx,
            'answer': answer,
        }

        paragraph_item['qas'].append(qas_item)
        
        data_item = {
            'title': 'synthetic',
            'paragraphs': [paragraph_item],
        }

        synthetic['data'].append(data_item)
        
        idx += 1
```

```python
pp.pprint(synthetic)
```

```python
with open("test/synthetic_task2.json", "w") as f:
    json.dump(synthetic, f)
```

```python
_id = []
_label = []
synthetic_data = synthetic['data']
for item in synthetic_data:
    paragraphs = item['paragraphs']
    for paragraph in paragraphs:
        qas = paragraph['qas']
        for q in qas:
            _id.append(q['id'])
            _label.append(q['answer'])
```

```python
golden = {}
for num, i in enumerate(_id):
    golden[i] = _label[num]

pp.pprint(golden)
```

```python
with open("test/task2_golden.json", "w") as f:
    json.dump(golden, f)
```

```python
with open("../test/synthetic_task2.json", "w") as f:
    json.dump(synthetic, f)

with open("../test/task2_golden.json", "w") as f:
    json.dump(golden, f)
```
